```python
import os
import subprocess
from urllib import parse as urlparse
#!yt-dlp --output "part3.mp4" --merge-output-format mp4 "https://www.youtube.com/watch?v=ldVXUfFAGIc" --downloader ffmpeg

# detect filesize ? keep: remove

# Youtube video url & corresponding timestamps
# Tuples are in format 'start','end' in seconds
def time_to_seconds(timestamp):
    minutes, seconds = map(int, timestamp.split(':'))
    total_seconds = (minutes * 60) + seconds
    return total_seconds


file_dict = {
    'part1.mp4': [
        ('0:06', '0:09'),    # 0:06 to 0:08
        ('1:20', '1:30'),    # 1:25 to 1:31
        ('1:57', '2:01'),    # 1:57 to 2:01
        ('2:44', '2:49'),    # 2:44 to 2:49
        ('2:56', '2:59'),    # 2:56 to 2:59
        ('3:45', '3:47'),    # 3:45 to 3:47
        ('4:03', '4:06'),    # 4:03 to 4:06
        ('5:03', '5:17')     # 5:03 to 5:17
    ],
    'part2.mp4': [
        ('0:20', '0:25'),    # 0:20 to 0:25
        ('1:14', '1:19'),    # 1:14 to 1:19
        ('3:08', '3:10'),    # 3:08 to 3:10
        ('6:22', '6:30'),    # 6:22 to 6:30
        ('7:00', '7:03'),    # 7:00 to 7:03
        ('7:08', '7:11'),    # 7:08 to 7:11
        ('7:27', '7:30'),    # 7:27 to 7:30
        ('7:34', '7:39'),    # 7:34 to 7:39
        ('7:51', '7:53')     # 7:51 to 7:53
    ],
    'part3.mp4': [
        ('0:30', '0:33'),    # 0:30 to 0:33
        ('0:39', '0:41')     # 0:39 to 0:41
    ]

}


def create_clips():
    OUTPUTCOUNT = 0
    total = 0
    video_files = []
    for file_name, intervals in file_dict.items(): 
        print(f'{file_name}')
        for interval in intervals: 
            start, end = interval
            if isinstance(start, str):
                start = time_to_seconds(start)
            if isinstance(end, str):
                end = time_to_seconds(end)
            time = end - start
            cmd = f'ffmpeg -i {file_name} -ss {start} -t {time} -c:v copy -c:a copy output{OUTPUTCOUNT}.mp4'
            subprocess.call(cmd, shell=True)
            video_files.append(f'output{OUTPUTCOUNT}.mp4')
            OUTPUTCOUNT += 1
    create_feeder_txt(video_files)


def create_feeder_txt(video_files):
    input_file_content = '\n'.join([f"file '{file}'" for file in video_files])
    with open('input.txt', 'w') as input_file:
        input_file.write(input_file_content)
    merge()

def merge():
    clip_merge_cmd = f'ffmpeg -f concat -safe 0 -i input.txt -c:v copy -c:a copy output.mp4'
    subprocess.call(clip_merge_cmd, shell=True)

def run(): 
    create_clips()

run()
#!ffmpeg -i output0.mp4 -i output2.mp4 -filter_complex "[0:v][0:a][1:v][1:a]concat=n=2:v=1:a=1[vout][aout]" -map "[vout]" -map "[aout]" concat.mp4


```
